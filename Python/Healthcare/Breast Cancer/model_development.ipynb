Load Packages

```python
import os
import sys
import numpy as np
import pandas as pd

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import joblib
```

Pandas display configuration

```python
pd.set_option("display.max_columns", 50)
pd.set_option("display.max_info_rows", 50)
```

Load dataset

```python
cancer = pd.read_csv("data.csv")
cancer.head(10)
```

Dataset Info

```python
cancer.info()
```

Missing values

```python
cancer.isna().sum()
```

Drop unused columns

```python
unused_col = ["id", "Unnamed: 32"]
cancer_copy = cancer.copy(deep=True)
new_cancer = cancer_copy.drop(columns=["id", "Unnamed: 32"])
new_cancer.head()
```

```python
new_cancer.isna().sum()
```

Split dataset

```python
X = new_cancer.iloc[:, 1:]
y = new_cancer.iloc[:, 0]

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, stratify=y, random_state=25)
Xtrain.shape, Xtest.shape
```

Build and Fit Model Pipeline

```python
classifier = KNeighborsClassifier()
model = Pipeline(steps=[("scaler", StandardScaler()), ("classifier", classifier)])
model.fit(Xtrain, ytrain)
```

```python
# Evaluate model
train_acc = model.score(Xtrain, ytrain)
test_acc = model.score(Xtest, ytest)
print("Train acc: ", train_acc)
print("Test acc: ", test_acc)
```

Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
help(classifier)
```

```python
# Get tunable parameters
classifier.get_params()
```

```python
# grid search parameters
param_grid = {"n_neighbors": [3, 5, 7, 9, 11],
              "weights": ["uniform", "distance"]}

# grid search model
grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid)

# create pipeline
grid_pipe = Pipeline(steps=[("scaler", StandardScaler()), ("grid_search", grid_search)])

# Fit gridsearch pipeline model
grid_pipe.fit(Xtrain, ytrain)

# grid model with no pipeline
grid_model = grid_pipe.named_steps["grid_search"]

# grid search results
cv_results = grid_model.cv_results_

```

```python
# model with best parameters
grid_model.best_params_
```

```python
# Additional considerations for choosing production model
result_columns = ["mean_test_score", "std_test_score", "rank_test_score"] + ["mean_fit_time", "std_fit_time", "mean_score_time", "std_score_time"] + ["param_" + param for param in param_grid.keys()]
results_data = pd.DataFrame(cv_results)
results_data.loc[:, result_columns].sort_values(by=["mean_test_score", "std_test_score","mean_fit_time", "std_fit_time"],
                                                ascending=[False, True, True, True])
```

Pipeline Model for Best Estimator

```python
best_estimator = grid_model.best_estimator_
estimator_pipeline = Pipeline(steps=[("scaler", StandardScaler()), ("classifier", best_estimator)])
estimator_pipeline.fit(Xtrain, ytrain)
```

```python
estimator_pipeline.score(Xtest, ytest)
```

Save Model Pipeline

```python
joblib.dump(value=estimator_pipeline, filename="knn_pipeline.pkl")

model_features = Xtrain.columns.tolist()
joblib.dump(model_features, "model_features.pkl")
```

Load Saved Model

```python
loaded_model = joblib.load("knn_pipeline.pkl")
loaded_model.score(Xtrain, ytrain)
```

```python
loaded_model.predict(Xtest)
```
